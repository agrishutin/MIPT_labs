```python
x = a * a
y = T * T * a

k, b = np.polyfit(x, y, 1)

grid = np.linspace(x.min(), x.max(), 500)

ax = plt.figure(figsize=(15, 10)).add_subplot(1, 1, 1)
ax.scatter(a * a, T * T * a, edgecolor='b', label='Практические данные')
ax.plot(x, k * x + b, color='red', label='Наилучшее приближение')

minor_xticks = np.arange(-0, 2100, 50)
major_xticks = np.arange(-0, 2100, 500)
minor_yticks = np.arange(30, 120, 1)
major_yticks = np.arange(30, 120, 10)

ax.set_xticks(major_xticks)
ax.set_xticks(minor_xticks, minor=True)
ax.set_yticks(major_yticks)
ax.set_yticks(minor_yticks, minor=True)

ax.grid(which='minor', alpha=0.4, ls=':')
ax.grid(which='major', alpha=0.6, ls=':')

plt.xlabel("a")
plt.ylabel("$T^2a$")
plt.legend()
# plt.show()

plt.savefig('T2a(a2).png')

print("K = {k}, b = {b}, g = {g}".format(k=k, b=b, g= 4 * pi * pi / k / 100))
```
