```python
harmonics_nums = np.array([1, 2, 3, 4, 5, 10])

temperatures = np.array([22.2, 30.2, 40.0, 50.0, 55.1])

resonance_freq = np.array([[204, 447, 659, 872, 1090, 2170],
                           [202, 451, 669, 883, 1102, 2195],
                           [205, 459, 676, 897, 1118, 2231],
                           [210, 467, 686, 915, 1140, 2280],
                           [215, 470, 697, 930, 1144, 2285]])
```

```python
measures = []
for t, resonances in zip(temperatures, resonance_freq):
    for num, resonance_freq in zip(harmonics_nums, resonances):
        measures.append([t, num, resonance_freq])

measures = pd.DataFrame(data=measures, 
                        columns=[r"Температура, $^\circ C$", 
                                 u"№ гармоники",
                                 u"частота, Гц"])
measures
```

```python
plt.figure(figsize=(18, 40))
grid = np.linspace(0, 10, 1000)

for t, resonances in zip(temperatures, resonance_freq):    
    k, b = np.polyfit(harmonics_nums, resonances, 1)
    
    plt.scatter(harmonics_nums,
                 resonances,
                 label="{} C".format(t))
    
    plt.plot(grid, 
             k * grid + b,
             label="{} C".format(t))

plt.legend(loc='best', fontsize=16)
plt.show()
```
