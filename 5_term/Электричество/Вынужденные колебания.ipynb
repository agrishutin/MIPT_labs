# Определение зависимости резонансной частоты от емкости

```python
C = np.arange(1, 11) * 10 ** (-3)
freq = np.array([11.6,
                 8.45,
                 6.9,
                 5.89,
                 5.3,
                 4.9,
                 4.5,
                 4.2,
                 3.95,
                 3.76])
n = len(C)
df = pd.DataFrame(data=np.hstack((C.reshape(n, 1) * 10 ** 3, freq.reshape(n, 1))),
                  columns=[r"C, нФ",
                           r"$f_p, кГц$"])
df
```

Резонансная частота колебаний контура(она же частота свободных колебаний) может быть получена по формуле:
$$
    \omega^2 = \frac{1}{LC}
$$
Значит, индуктивность катушки можно найти как коэффициент наклона графика $F(C)$, где
$$
    F(C) = \frac{1}{(2 \pi f_p)^2}.
$$

```python
x = C
y = 1 / ((2 * np.pi * freq) ** 2)

k = np.mean(x * y) / np.mean(x ** 2)
b = 0

plt.figure(figsize=(20, 10))
plt.plot(C, 1 / ((2 * np.pi * freq) ** 2), '-o')

C1 = np.arange(0, 11) * 10 ** (-3)
plt.plot(C1, k * C1 + b, 'r')
plt.scatter(0, 0)
plt.show()
```
