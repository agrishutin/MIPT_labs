# Теория.
   По своим свойствам жидкости сходны как с газами, так и с твердыми телами. Подобно газам, жидкости принимают форму сосуда, в котором они находятся. Подобно твердым телам, они обладают сравнительно большой плотностью, с трудом поддаются сжатию.
   Двойственный характер свойств жидкостей связан с особенностями движения их молекул. В газах молекулы движутся хаотично, в их расположении отсутствует порядок. В кристаллических твердых телах частицы колеблются около определенных положений равновесия --- узлов кристаллической решетки. В жидкостях, как и в кристаллах, каждая молекула находится в потенциальной яме электрического поля, создаваемого окружающими молекулами. Молекулы колеблются со средней частотой, близкой к частоте колебаний атомов в кристаллических телах ( ~$10^{12}$  Гц), и с амплитудой, определяемой размерами объема, предоставленного ей соседниим молекулами. Глубина потенциальной ямы в жидкостях больше средней кинетической энергии колеблющейся молекулы, поэтому молекулы колеблются вокруг более или менее стабильных положений равновесия.

#### Чтобы перейти в новое состояние, молекула должно преодолеть участки с большой потенциальной энергией -- с энергией, превышающей среднюю тепловую энергию молекул. Для этого тепловая энергия молекул должна -- вследствие флуктуации -- увеличиться на некоторую величину $ W $ , называемую * энергией активации *.

#### Количество молекул, имеющих энергии больше $ W $,  в соответствии с формулой Больцмана, экспоненциально зависит от $ W $. Температурная зависимость вязкости жидкости выражается формулой (12): 
$$
\eta \sim Ae^{{W}/_{kT}}
$$

## Метод Стокса:
На всякое тело, двигающееся в вязкой жидкости, действует сила сопротивления. Стокс выявил, что при ламинарном обтекании гарика безграничной жидкостью сила сопротивления $ F $ определяется формулой (13):
$$
F = 6 \pi \eta r v
$$
где $ \eta $ -- вязкость жидкости, $ v $ -- скорость шарика, $ r $ - его радиус.

### Рассмотрим свободное падение шарика в вязкой жидкости:
На него действуют 3 силы: сила тяжести, сила Архимеда и сила вязкости, зависящая от скорости.
Найдем уравнение движения шарика в жидкости. Из второго закона Ньютона:
$$
Vg(\rho - \rho_ж) - 6 \pi \eta r v = V\rho\frac{dv}{dt}
$$
где $ V $ -- объем шарика, $ \rho $ -- его плотность, $ \rho_ж $ -- плотность жидкости, $ g $ -- ускорение свободного падения.

Решаем уравнение и находим (15):

$$
v(t) = v_{уст} - \big[ v_{уст} - v(0) \big] e^{\frac{-t}{\tau}}
$$
где $ v(0) $ -- скорость шарика в начальный момент, 
$ v_{уст} = \frac{Vg(\rho - \rho_ж)}{6 \pi \eta r} = 
\frac{2}{9} g r^{2}\frac{(\rho - \rho_ж)}{\eta}$, 

$ \tau = \frac{V\rho}{6 \pi \eta r} = 
\frac{2}{9}\frac{r^{2}\rho}{\eta} $

## Формула для нахождения вязкости:
$$
\eta = \frac{2}{9}gr^{2}\frac{\rho - \rho_ж}{v_{уст}}
$$

# Указания
Описанная методика определения вязкости основана на формуле Стокса и правильна лишь в том случае, если выполнены предположения, сделанные при выводе этой формулы. Самым надежным способом проверки теории является следующий: Шарики с разными радиусами движутся в жидкости с разными скоростями и разными временами релаксации. Если во всем диапазоне встречающихся в работе скоростей и времен релаксации вычисленные по формуле $ \eta $ оказываются одинаковыми, то формула на F правильно передает зависимость сил от радиуса шарика. Зависимость(или независимость) $ \eta $ от $ r $ служит чувствительным индикатором  правильности теории и надежности эксперимента.

Результаты опыта имеет смысл обрабатывать лишь в том случае, если значения $ \eta $ не обнаруживают систематической зависимости от $ r $. Если такая зависимость наблюдается, то чаще всего это связано с влиянием стенок сосуда.  В этом случае вместо формулы выше стоит использовать более точную формулу:

$$ 
\eta = \frac{2}{9}gr^{2}\frac{\rho - \rho_ж}{\big[ 1 + 2,4(^{r}/_{R}) \big]v_{уст}}
$$
где $ R $ -- радиус сосуда. Для небольших шариков отличие формул лежит в пределах точности эксперимента и может не приниматься во внимание. (последняя формула верна для шариков, движущихся вблизи осевой линии сосуда, при условии r << R)

При выводе формулы Стокса предполагалось, что обтекание шарика жидкостью имеет ламинарный характер.
Как известно, характер обтекания определяется значением числа Рейнольдса $Re = ^{vr\rho_{ж}}/_{\eta} $

Обтекание является ламинарным лишь при не очень больших значениях $Re$ ( $< 10 $ ). По результатам опыта следует вычислить числа Рейнольдса для разных размеров шариков и различных температур жидкости. Полезно также вычислить время релаксации $\tau$  и путь релаксации $ S $, который может быть найден посредством интегрирования (15).

Полагаем, для простоты, $ v(0) = 0 $(что обычно выполняется с достаточной точностью) и получаем (19):
$$
S = v_{уст}\tau\Big(\frac{t}{\tau} - 1 + e^{\frac{-t}{\tau}} \Big)
$$

Из (19) легко найти, что $ S >> \tau v_{уст}$ при $ t >> \tau $.
