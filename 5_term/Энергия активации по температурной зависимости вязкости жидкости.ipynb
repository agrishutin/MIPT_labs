# Теория

#### Чтобы перейти в новое состояние, молекула должно преодолеть участки с большой потенциальной энергией -- с энергией, превышающей среднюю тепловую энергию молекул. Для этого тепловая энергия молекул должна -- вследствие флуктуации -- увеличиться на некоторую величину $ W $ , называемую * энергией активации *.

#### Количество молекул, имеющих энергии больше $ W $,  в соответствии с формулой Больцмана, экспоненциально зависит от $ W $. Температурная зависимость вязкости жидкости выражается формулой (12): 
$$
\eta \sim Ae^{{W}/_{kT}}
$$

## Метод Стокса:
На всякое тело, двигающееся в вязкой жидкости, действует сила сопротивления. Стокс выявил, что при ламинарном обтекании гарика безграничной жидкостью сила сопротивления $ F $ определяется формулой (13):
$$
F = 6 \pi \eta r v
$$
где $ \eta $ -- вязкость жидкости, $ v $ -- скорость шарика, $ r $ - его радиус.

### Рассмотрим свободное падение шарика в вязкой жидкости:
На него действуют 3 силы: сила тяжести, сила Архимеда и сила вязкости, зависящая от скорости.
Найдем уравнение движения шарика в жидкости. Из второго закона Ньютона:
$$
Vg(\rho - \rho_ж) - 6 \pi \eta r v = V\rho\frac{dv}{dt}
$$
где $ V $ -- объем шарика, $ \rho $ -- его плотность, $ \rho_ж $ -- плотность жидкости, $ g $ -- ускорение свободного падения.

Решаем уравнение и находим (15):

$$
v(t) = v_{уст} - \big[ v_{уст} - v(0) \big] e^{\frac{-t}{\tau}}
$$
где $ v(0) $ -- скорость шарика в начальный момент, 
$ v_{уст} = \frac{Vg(\rho - \rho_ж)}{6 \pi \eta r} = 
\frac{2}{9} g r^{2}\frac{(\rho - \rho_ж)}{\eta}$, 

$ \tau = \frac{V\rho}{6 \pi \eta r} = 
\frac{2}{9}\frac{r^{2}\rho}{\eta} $

## Формула для нахождения вязкости:
$$
\eta = \frac{2}{9}gr^{2}\frac{\rho - \rho_ж}{v_{уст}}
$$

# Указания
Описанная методика определения вязкости основана на формуле Стокса и правильна лишь в том случае, если выполнены предположения, сделанные при выводе этой формулы. Самым надежным способом проверки теории является следующий: Шарики с разными радиусами движутся в жидкости с разными скоростями и разными временами релаксации. Если во всем диапазоне встречающихся в работе скоростей и времен релаксации вычисленные по формуле $ \eta $ оказываются одинаковыми, то формула на F правильно передает зависимость сил от радиуса шарика. Зависимость(или независимость) $ \eta $ от $ r $ служит чувствительным индикатором  правильности теории и надежности эксперимента.

Результаты опыта имеет смысл обрабатывать лишь в том случае, если значения $ \eta $ не обнаруживают систематической зависимости от $ r $. Если такая зависимость наблюдается, то чаще всего это связано с влиянием стенок сосуда.  В этом случае вместо формулы выше стоит использовать более точную формулу:

$$ 
\eta = \frac{2}{9}gr^{2}\frac{\rho - \rho_ж}{\big[ 1 + 2,4(^{r}/_{R}) \big]v_{уст}}
$$
где $ R $ -- радиус сосуда. Для небольших шариков отличие формул лежит в пределах точности эксперимента и может не приниматься во внимание. (последняя формула верна для шариков, движущихся вблизи осевой линии сосуда, при условии r << R)

При выводе формулы Стокса предполагалось, что обтекание шарика жидкостью имеет ламинарный характер.
Как известно, характер обтекания определяется значением числа Рейнольдса $Re = ^{vr\rho_{ж}}/_{\eta} $

Обтекание является ламинарным лишь при не очень больших значениях $Re$ ( $< 10 $ ). По результатам опыта следует вычислить числа Рейнольдса для разных размеров шариков и различных температур жидкости. Полезно также вычислить время релаксации $\tau$  и путь релаксации $ S $, который может быть найден посредством интегрирования (15).

Полагаем, для простоты, $ v(0) = 0 $(что обычно выполняется с достаточной точностью) и получаем (19):
$$
S = v_{уст}\tau\Big(\frac{t}{\tau} - 1 + e^{\frac{-t}{\tau}} \Big)
$$

Из (19) легко найти, что $ S >> \tau v_{уст}$ при $ t >> \tau $.

```python
# Все измеряем в см, сек, г

rho_glass = 2.5         # плотность стекла, г/см^3
rho_steel = 7.8         # плотность стали, г/см^3
g =         9.8 * 100   # ускорение св.пад. cм/c^2
L =         10          # расстояние между рисками, см
k_bolc    = 1.380 * 10 ** (-23) # постоянная Больцмана

def celsium_to_kelvin(t):
    return t + 273.15

def get_glycerol_density(t):
    r25 = 1.258 # г/см^3
    r35 = 1.255 # г/см^3
    k = (r35 - r25) / (35 - 25)
    
    return r25 + k * (t - 25)

def volume(r):
    return (4 / 3) * np.pi * (r ** 3)

def get_v(t):
    return L / t

def get_eta(r, rho, t, T):
    v_ust = get_v(t)
    
    return (2 / 9) * g * (r ** 2) * ((rho - get_glycerol_density(T)) / v_ust)

def get_tau(r, rho, t, T):
    eta = get_eta(r, rho, t, T)
    return (2 / 9) * (r ** 2 * rho) / (eta)

def get_s(r, rho, t, T):
    v = get_v(t)
    tau = get_tau(r, rho, t, T)
    
    return v * tau

def get_Reinolds(r, rho, t, T):
    eta = get_eta(r, rho, t, T)
    v = get_v(t)
    
    return v * get_glycerol_density(T) * r / eta

def get_activation_energy(plot_k):
    return k_bolc * plot_k

# погрешности

sigma_r = 0.002  # см
sigma_l = 0.1    # см
sigma_t = 0.25   # с
sigma_T = 0.1    # К

def get_sigma_v(t):
    return math.sqrt(sigma_l ** 2 / t ** 2 + (d ** 2 * sigma_t ** 2) / t ** 4)

def get_sigma_eta(r, rho, t, T):
    v_ust = get_v(t)
    sigma_v = get_sigma_v(t)
    return (2 / 9) * math.sqrt(((2 * g * r * (rho - get_glycerol_density(T)) / v_ust) * sigma_r) ** 2 + 
                               (g * (r ** 2) * (rho - get_glycerol_density(T)) / (v_ust ** 2) * sigma_v) ** 2)

def get_mean_sigma_eta(eta_sigmas):
    return math.sqrt(np.sum(eta_sigmas * eta_sigmas))

def get_ln_mean_sigma_eta(etas, eta_sigmas):
    mean_sigma_eta = get_mean_sigma_eta(eta_sigmas)
    
    return mean_sigma_eta / np.mean(etas)

def get_sigma_inversed_T(T):
    return sigma_T / T ** 2
```

```python
# цена деления - 0.02 мм
diameters = np.array([0,
                      1.2,
                      2,
                      2,
                      2.2,
                      1.9,
                      2,
                      2.1,
                      2.1,
                      2.2,
                      0.72,
                      0.6,
                      0.72,
                      0.9,
                      0.62,
                      0.59,
                      0.56,
                      0.59,
                      0.71])

for i, old_val in enumerate(diameters):
    diameters[i] = old_val // 1 + (old_val % 1) / 2

diameters = diameters / 10
diameters_df =  pd.DataFrame(data=diameters[1:],
                         columns=["Диаметр шарика, cм"])
diameters_df
```

```python
# использовавшиеся температуры, градусы Цельсия
temperatures_celsium = np.array([24.9, 
                                 30.5, 
                                 35.35, 
                                 42.4, 
                                 45.3])

temperatures = np.array([celsium_to_kelvin(t) for t in temperatures_celsium])

# Массив номеров шариков на каждом замере. Первые 9 стеклянные.
numbers = [[1, 10, 2],
           [3, 4, 11, 12],
           [5, 6, 13, 14],
           [7, 8, 15, 16],
           [9, 17, 18]]

# время, за которое шарик пролетал расстояние между двумя последними рисками
times = [[12.75, 15.28, 12.25],
         [10.78, 10.34, 14.43, 14.93],
         [7.81, 7.22, 10.44, 6.19],
         [4.65, 4.59, 11.47, 10.69],
         [4.3, 6.53, 7.28]]
```

```python
all_measures = []
mean_measures = []

# all measures
coeffs = np.array([1,
                   1,
                   1,
                   1,
                   1,
                   1,
                   1,
                   10 ** 3,
                   10 ** 3,
                   10 ** 3,
                   10 ** 2])

columns = [r"$T, K$", 
           r"$r, мм$", 
           r"$\rho, \frac{г}{см^3}$", 
           r"$t, сек$",
           r"$v, \frac{см}{c^2}$",
           r"$\eta, \frac{кг}{м \cdot сек}$",
           r"$Re$",
           r"$\tau, с \cdot 10^{-3}$",
           r"$S, см \cdot 10^{-3}$",
           r"$\sigma(v) \cdot 10^{-3} $",
           r"$\sigma(\eta), \frac{кг}{м \cdot сек} \cdot 10^{-2}$"]

# mean measures
mean_columns = [r"$T, K$",
                r"$\overline{\eta}, \frac{кг}{м \cdot сек}$",
                r"$\sigma(\overline{\eta}), \frac{кг}{м \cdot сек}$",
                r"$\ln{\overline{\eta}}$",
                r"$\sigma(\ln{\overline{\eta}})$",
                r"$\frac{1}{T}, \frac{10^{-3}}{K}$",
                r"$\sigma(\frac{1}{T}), \frac{10^{-5}}{K}$"]

mean_coeffs = np.array([1,
                        1,
                        1,
                        1,
                        1,
                        10 ** 3,
                        10 ** 5])


for i, T in enumerate(temperatures):
    cur_etas = []
    cur_sigma_etas = []
    
    for j, number in enumerate(numbers[i]):
        # data
        d = diameters[number]
        r = d / 2
        t = times[i][j]
        rho = rho_glass if number <= 9 else rho_steel
        
        # calculations
        v = get_v(t)
        eta = get_eta(r, rho, t, T)
        Re = get_Reinolds(r, rho, t, T)
        tau = get_tau(r, rho, t, T)
        S = get_s(r, rho, t, T)
        sigma_v = get_sigma_v(t)
        sigma_eta = get_sigma_eta(r, rho, t, T)
        
        # new entry for global table
        cur_row = np.array([T, r, rho, t, v, eta, Re, tau, S, sigma_v, sigma_eta]) * coeffs
        all_measures.append(cur_row)
        
        cur_etas.append(eta)
        cur_sigma_etas.append(sigma_eta)
        
    etas = np.array(cur_etas)
    sigma_etas = np.array(cur_sigma_etas)
    
    cur_mean_measure = np.array([T, 
                                 np.mean(cur_etas),
                                 get_mean_sigma_eta(sigma_etas),
                                 np.log(np.mean(cur_etas)), 
                                 get_ln_mean_sigma_eta(etas, sigma_etas),
                                 1 / T,
                                 get_sigma_inversed_T(T)]) * mean_coeffs
    mean_measures.append(cur_mean_measure)

all_measures = np.array(all_measures)
mean_measures = np.array(mean_measures)
```

## Найдем энергию активации
Как известно, $ \ln{\eta} = \frac{W}{kT} $. Значит, 
$ W = k \ln{\eta}T = k \frac{\Delta{\ln{\overline{\eta}}}}{\Delta{\frac{1}{T}}} \approx k \cdot k_{графика} \approx 6.928 \cdot 10 ^{-20} Дж$, при справочном значении $W = (8.341 \pm 0.087) \cdot 10 ^ {-20} Дж $.
